```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import numpy as np
```

```python
# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)
df.sample(10)
```

```python
## Write your code here
language = df.loc[:, ['LanguageHaveWorkedWith']]
language.sample(10)
```

```python
# Step 1: Explode the Language combinations into separate rows
language['LanguageHaveWorkedWith'] = language['LanguageHaveWorkedWith'].str.split(';')
language_exploded = language.explode('LanguageHaveWorkedWith')

# Step 2: Count language popularity
language_popularity = language_exploded['LanguageHaveWorkedWith'].value_counts().reset_index()
language_popularity.columns = ['Language', 'TotalCount']

# Display the results
print("Exploded DataFrame:")
print(language_exploded)
print("\nLanguage Popularity:")
print(language_popularity)
```

```python
top_10_languages = language_popularity.head(10)
top_10_languages.to_csv('/Users/longhoachung/top_10_languages.csv', index=False)
```

```python
## Write your code here
country = df.loc[:, ['Country']]
country.sample(10)
```

```python
country.describe()
```

```python
country.isnull().sum()
```

```python
### Scatter plot section Lab 16 has the codes to clean up country data
country = country.dropna()
```

```python
country.isnull().sum()
```

```python
country['Country'].value_counts()
```

```python
country_list=country['Country'].unique()
print(len(country_list), '\n')
print(country_list)
```

```python
# Turn the Country into str first, then strip two ends then in the middle with regex
country['Country'] = country['Country'].astype(str)
country['Country'] = country['Country'].str.strip()
country['Country'] = country['Country'].str.replace('Republic Of Korea', '', regex=False)
country['Country'] = country['Country'].str.replace("Democratic People'S Republic Of Korea", '', regex=False)
country['Country'] = country['Country'].str.replace("Democratic Republic Of The Congo", '', regex=False)
country['Country'] = country['Country'].str.replace("Micronesia, Federated States", '', regex=False)
country['Country'] = country['Country'].str.replace("Democratic People'S ", '', regex=False)
country['Country'] = country['Country'].str.replace(',', '', regex=False)
country['Country'] = country['Country'].str.title()
country['Country'] = country['Country'].replace('United Kingdom Of Great Britain And Northern Ireland', 'UK and IRE')
country['Country'] = country['Country'].replace('United States Of America', 'USA')
country['Country'] = country['Country'].replace('Iran Islamic Republic Of...', 'Iran')
country['Country'] = country['Country'].replace('United Republic Of Tanzania', 'Tanzania')
country['Country'] = country['Country'].replace("Lao People'S Democratic Republic", 'Lao')
country['Country'] = country['Country'].replace('Central African Republic', 'Central Africa')
country['Country'] = country['Country'].replace('Congo Republic Of The...', 'Congo')
country['Country'] = country['Country'].replace('Micronesia, Federated States Of', 'Micronesia')
country['Country'] = country['Country'].replace('Libyan Arab Jamahiriya', 'Libya')
country['Country'] = country['Country'].replace('Brunei Darussalam', 'Brunei')
country['Country'] = country['Country'].replace('Russian Federation', 'Russia')
country['Country'] = country['Country'].replace('Venezuela Bolivarian Republic Of...', 'Venezuela')
```

```python
country_list = country['Country'].value_counts()
```

```python
country_list = country['Country'].value_counts().reset_index()
country_list.columns = ['Country', 'TotalCount']
```

```python
country_list.to_csv('/Users/longhoachung/countries.csv', index=False)
```

```python
age_edu = df.loc[:, ['Age', 'EdLevel']]
age_edu.sample(10)
```

```python
age_edu.isnull().sum()
```

```python
age_edu.loc[:, 'EdLevel'] = age_edu['EdLevel'].fillna('No response')
```

```python
age_edu.isnull().sum()
```

```python
age_edu.sample(10)
```

```python
grouped = age_edu.groupby(["Age", "EdLevel"]).size().reset_index(name="Count")
```

```python
grouped
```

```python
grouped.loc[:, 'EdLevel'] = grouped['EdLevel'].astype(str)
```

```python
grouped.loc[grouped['EdLevel'] == "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)", 'EdLevel'] = "Masters degree (M.A., M.S., M.Eng., MBA, etc.)"
```

```python
grouped.loc[grouped['EdLevel'] == "Bachelor’s degree (B.A., B.S., B.Eng., etc.)", 'EdLevel'] = "Bachelors degree (B.A., B.S., B.Eng., etc.)"
```

```python
grouped['EdLevel'].value_counts()
```

```python
grouped.sample(10)
```

```python
grouped.groupby('Age').sum()
```

```python
grouped.info()
```

```python
edu_background = grouped.groupby('Age')['Count'].sum()
```

```python
edu_background
```

```python
grouped['PercentInAgeGroup'] = grouped['Count'] / grouped['Age'].map(edu_background) * 100
```
